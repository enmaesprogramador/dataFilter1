```python
import pandas as pd 
import openpyxl 

from pandasql import sqldf
# Funcion lambda para correr querrys mas facil
pysqlc = lambda q: sqldf(q, globals())
```

```python
# Extraccion de los datos
ar = pd.read_excel('country_tables.xlsx')
```

# Creacion de los nuevos DataFrames

Solo escogiendo los que tienen una poblacion MAYOR a 50,000,000 y MENOR a 60,000,000 y con una Superficie MAYOR a 3,000,000 y menor a 3,000,000

Paises con poblacion y superficie mayores

```python
# Se crea el arreglo de columnas
# Luego se crea un nuevo data frame con los datos de esas columnas cumpliendo las condiciones
column_data = ['Nombre', 'Poblacion', 'Superficie']
ar_new_data = ar[ar['Poblacion'] >= 50000000][ar['Superficie'] >= 3000000][column_data]
```

```python
# Despliegue del nuevo DataFrame
ar_new_data
```

Paises con superficie y poblacion menor

```python
ar_new_data2 = ar[ar['Poblacion'] <= 60000000][ar['Superficie'] <= 3000000][column_data]
```

```python
ar_new_data2
```

# Consulta SQL y extraccion de datos

```python
full_df1 = """
SELECT 
    Nombre,
    Poblacion,
    Superficie
FROM
    ar_new_data
UNION ALL
SELECT 
    Nombre,
    Poblacion,
    Superficie
FROM
    ar_new_data2
ORDER BY 
    Poblacion, Superficie DESC
"""

full_df_Export1 = pysqlc(full_df1)
```

```python
# Conversion del nuevo archivo a uno excel
full_df_Export1.to_excel('fulldataFrame.xlsx')
```
